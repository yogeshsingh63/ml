```python
import time
start_time = time.time()
```

```python
import torch
import numpy as np
```

```python
# from direct data
data = [[1,2], [3,4]]
x_data = torch.tensor(data)
x_data
```

```python
# from a numpy array
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
x_np
```

```python
# from another tensors
x_ones = torch.ones_like(x_data) # Retains shape, datatype of the x_data
print(f"Ones Tensor:\n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) #overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# With random or constant values

shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor:\n {rand_tensor} \n")
print(f"Ones Tensor:\n {ones_tensor} \n")
print(f"Zeros Tensor:\n {zeros_tensor} \n")
```

```python
# Tensor attributes

tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Tensor operations ( Moving to another device if available)

device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device)
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Numpy like indexing and slicing

t1 = torch.rand(4,4)
t2 = torch.rand(4,4)
t3 = torch.rand(4,4)

t1[2,:] = 0
print(t1)
print(t2)
print(t3)
```

```python
# concat
t4 = torch.cat([t1, t2 , t3], dim=-1)
print(t4)
```

```python
# Multiplying tensors

print(f"tensor.mul(tensor):\n {t1.mul(t2)}\n")
#Alternative syntax
print(f"tensor*tensor:\n {tensor*tensor} \n")
```

```python
# Matrix multiplication

print(f"t1.matmul(t2.T):\n {t1.matmul(t2.T)}\n")
#Alternative Syntax
print(f"t1 @ t2.T:\n {t1 @ t2.T}\n")
```

```python
print(t1)
t1.add_(5)
print(t1)
```

```python
# Bridge with numpy
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```

```python
end_time = time.time()
elpased_time = end_time - start_time
print(f"Total runtime: {elpased_time}")
```
