```python
import time
start_time = time.time()
```

```python
# Defining a Neural Network
import torch
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):

  def __init__(self):
    super().__init__()

    self.conv1 = nn.Conv2d(1, 6, 5)
    self.conv2 = nn.Conv2d(6, 16, 5)

    self.fc1 = nn.Linear(16*5*5, 120)
    self.fc2 = nn.Linear(120, 84)
    self.fc3 = nn.Linear(84, 10)


  def forward(self, input):

    # convolution layer 1 , takes input (N*1*32*32) output tensor of (N*6*28*28)
    c1 = F.relu(self.conv1(input))

    #outputs a (N*6*14*14) tensor , by taking max of 2*2 blocks
    s2 = F.max_pool2d(c1, (2, 2))

    # conv layer 2 , tensor output (N*16*10*10)
    c3 = F.relu(self.conv2(s2))

    # again subsampling for max in 2*2 block , tensor output (N*16*5*5)
    s4 = F.max_pool2d(c3, 2)

    # flattens the per image tensor, output (N*400)
    s4 = torch.flatten(s4, 1)

    # fully connected Linear layer (400, 120)
    f5 = F.relu(self.fc1(s4))

    # fc 2 (120, 84)
    f6 = F.relu(self.fc2(f5))

    # fc3 final output logits , (84, 10)
    output = self.fc3(f6)

    return output
```

```python
net = Net()
print(net)
```

```python
params = list(net.parameters())
print(len(params))
print(params[0].size())
```

```python
input = torch.randn(1, 1, 32, 32)
out = net(input)
print(out)
```

```python
net.zero_grad()
out.backward(torch.randn(1, 10))
```

```python
end_time = time.time()
elapsed_time = end_time - start_time

elapsed_time
```
